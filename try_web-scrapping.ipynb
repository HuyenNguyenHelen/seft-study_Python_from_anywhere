```
import os
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup
url= "https://digital.library.unt.edu/ark:/67531/metadc849653/m2/1/high_res_d/RITTER-DISSERTATION-2016.pdf"
path= "C:\webscrapping\first_file.txt"
folder_location = os.open(path, "wb")
if not os.path.exists(folder_location): os.mkdir(folder_location)
response = requests.get (url)
soup = BeautifulSoup (response.text, "html.parser")
for link in soup.select ("a[href$='.pdf']"):
  filename = os.path.join (folder_location, link["href"].split('/')[-1])
  with open(filename, 'wb') as f:
    f.write (requests.get(urljoin(url, link['href'])).content)
```

```
from bs4 import BeautifulSoup
import requests
import re
url = "http://www.chessgames.com/perl/chesscollection?cid=1014492"
req = requests.get(url)
soup = BeautifulSoup(req.text, "lxml")
pages = soup.findAll('a', href=re.compile('.*chessgame\?.*'))
def download_file(url):
    path = url.split('/')[-1].split('?')[0]
    r = requests.get(url, stream=True)
    if r.status_code == 200:
        with open(path, 'wb') as f:
            for chunk in r:
                f.write(chunk)
host = 'http://www.chessgames.com'
for page in pages:
    url = host + page.get('href')
    req = requests.get(url)
    soup = BeautifulSoup(req.text, "lxml")
    file_link = soup.find('a',text=re.compile('.*download.*'))
    file_url = host + file_link.get('href')
    download_file(file_url)
```
