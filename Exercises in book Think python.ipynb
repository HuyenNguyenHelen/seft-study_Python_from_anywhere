```python


def do_twice(func, arg):
    """Runs a function twice.

    func: function object
    arg: argument passed to the function
    """
    func(arg)
    func(arg)


def print_twice(arg):
    """Prints the argument twice.

    arg: anything printable
    """
    print(arg)
    print(arg)


def do_four(func, arg):
    """Runs a function four times.

    func: function object
    arg: argument passed to the function
    """
    do_twice(func, arg)
    do_twice(func, arg)


do_twice(print_twice, 'spam')
print('')

do_four(print_twice, 'spam')
```

```python
# p.72, ex 6-2
def acker_func (m,n):
    if m==0:
        return n+1
    elif m>0 and n==0:
        return acker_func (m-1, 1)
    else:
        ext=acker_func(m, n-1)
        return acker_func (m-1,ext)
acker_func (5,6)
        
```
